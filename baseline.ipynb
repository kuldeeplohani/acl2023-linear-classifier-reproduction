```python
import time

datasets = ["ecthr_a", "ecthr_b", "scotus", "eurlex", "ledgar", "unfair_tos"]
methods = ["1vsrest", "thresholding", "cost_sensitive"]

start_time = time.time()

for data in datasets:
    for method in methods:
        print(f"\n===== Running {method.upper()} on {data.upper()} =====\n")
        cmd = f"bash run_experiments.sh {data} {method}"
        print(f"Command: {cmd}")
        !{cmd}
        print(f"\nFinished {method.upper()} on {data.upper()}!\n")

elapsed = (time.time() - start_time) / 60
print(f"\nAll runs completed in approximately {elapsed:.1f} minutes.")
```

```python
import os
import json
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === CONFIG ===
runs_dir = "runs"
records = []

# === PARSE ALL LOG FILES ===
for log_path in glob.glob(os.path.join(runs_dir, "*", "logs.json")):
    try:
        with open(log_path, "r") as f:
            data = json.load(f)

        folder = os.path.basename(os.path.dirname(log_path))
        dataset = data.get("config", {}).get("data_name", folder.split("_")[0])
        # detect model type
        if "bert_tuned" in folder:
            model = "BERT-Tuned"
        else:
            model = data.get("config", {}).get("linear_technique", "L2SVM").capitalize()

        # Handle list vs dict in "test" field
        test_entry = data.get("test", None)
        if isinstance(test_entry, list) and len(test_entry) > 0:
            test_metrics = test_entry[-1]
        elif isinstance(test_entry, dict):
            test_metrics = test_entry
        else:
            test_metrics = {}

        micro = test_metrics.get("Micro-F1")
        macro = test_metrics.get("Macro-F1")
        runtime = data.get("time", "N/A")

        records.append({
            "Dataset": dataset.upper(),
            "Model": model,
            "Micro-F1": micro,
            "Macro-F1": macro,
            "Runtime": runtime,
            "Folder": folder
        })

    except Exception as e:
        print(f" Error reading {log_path}: {e}")

# === CREATE DATAFRAME ===
df = pd.DataFrame(records).sort_values(["Dataset", "Model"]).reset_index(drop=True)
print("=== Summary Table ===")
display(df)

# === SAVE SUMMARY ===
df.to_csv("model_results_summary.csv", index=False)
print("\n Saved summary to model_results_summary.csv")

# === PLOT MICRO-F1 AND MACRO-F1 ===
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Dataset", y="Micro-F1", hue="Model")
plt.title("Micro-F1 Comparison across Models and Datasets")
plt.xticks(rotation=30, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Dataset", y="Macro-F1", hue="Model")
plt.title("Macro-F1 Comparison across Models and Datasets")
plt.xticks(rotation=30, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
